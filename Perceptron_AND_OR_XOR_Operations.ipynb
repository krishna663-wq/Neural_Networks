```python
import pandas as pd

# Create a DataFrame for the truth table
data = {'A': [0, 0, 1, 1],
        'B': [0, 1, 0, 1]}
df = pd.DataFrame(data)

# Calculate AND, OR, and XOR operations
df['AND'] = df['A'] & df['B']
df['OR'] = df['A'] | df['B']
df['XOR'] = df['A'] ^ df['B']


df
```

```python
from sklearn.linear_model import Perceptron
```

```python
obj1 = Perceptron()
obj2 = Perceptron()
obj3 = Perceptron()
```

```python
# fit the Perceptron.
obj1.fit(df[['A', 'B']], df['AND'])
obj2.fit(df[['A', 'B']], df['OR'])
obj3.fit(df[['A', 'B']], df['XOR'])
```

```python
obj1.coef_
```

```python
obj1.intercept_
```
